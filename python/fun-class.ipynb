```python
def fib2(n): #
	"""Docstring."""
	result = [] # local variable
	a,b = 0,1
	while a<n:
		result.append(a)
		a,b = b,a+b
	return result
```

The **default argument values** are **only created oncce**, when the function is defined.
```python
def f(a, L=[]):
    L.append(a)
    return L

print(f(1)) # [1]
print(f(2)) # [1, 2]

def f(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L

print(f(1)) # [1]
print(f(2)) # [2]
```
